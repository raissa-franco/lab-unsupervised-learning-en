# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here

# Column collinearity
# Since the columns 'Grocery' and 'Detergents_Paper' have high correlation, the decision made was to drop 'Detergents_Paper'.
# Dropping categorical columns: Region and Channel
wholesale=wholesale.drop(columns=['Detergents_Paper','Region','Channel'])
```

```python
# Outliers

num_cols = wholesale.select_dtypes(include='number').columns

# Create a copy of the dataframe to avoid modifying the original
wholesale_treated = wholesale.copy()

# Apply logarithmic transformation to handle outliers on selected numeric columns
wholesale_treated[num_cols] = wholesale_treated[num_cols].apply(lambda x: np.log1p(x))
```

Applying a logarithmic transformation to the numerical features is a common technique to reduce the impact of outliers and skewed distributions. This transformation compresses the range of large values while keeping the order of data points intact. By using log(x + 1), we avoid issues with zero or negative values, which are undefined in the logarithmic scale. This approach can help improve the performance of clustering algorithms by making the data distribution more balanced and less influenced by extreme values.

```python
# Checking distribution
plt.figure(figsize=(14, 8))
for i, column in enumerate(wholesale_treated.select_dtypes(include='number').columns):
    plt.subplot(2, 4, i + 1)
    sns.boxplot(y=wholesale_treated[column])
    plt.title(column)
plt.tight_layout()
plt.show()
```

I applied One-Hot Encoding to the Region and Channel columns because, although these columns are represented as integers, they are actually categorical variables without any inherent order. Using One-Hot Encoding transforms each category into a separate binary feature, which prevents the model from mistakenly interpreting these values as ordinal or continuous. This encoding helps ensure that machine learning algorithms treat these features appropriately, improving model performance and interpretability. Additionally, using drop_first=True avoids multicollinearity by removing one redundant category from each set of dummy variables.

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:

# Select columns to scale
cols_to_scale = [col for col in wholesale_treated.columns]

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling
customers_scale = wholesale_treated.copy()
customers_scale[cols_to_scale] = scaler.fit_transform(wholesale_treated[cols_to_scale])
```

```python
customers_scale.head()
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:

# Find the optimal number of clusters using the Elbow Method
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_scale)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)

labels = kmeans_2.predict(customers_scale)

clusters = kmeans_2.labels_.tolist()
```

```python
customers_scale['Label'] = clusters
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
eps_values = [0.3, 0.5, 0.8, 1.0, 1.2]

for eps in eps_values:
    dbscan = DBSCAN(eps=eps)
    labels = dbscan.fit_predict(customers_scale.drop(columns='Label'))
    customers_scale['labels_DBSCAN'] = labels
    
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=customers_scale['Grocery'], y=customers_scale['Milk'], hue=labels, palette='tab10')
    plt.title(f'DBSCAN Clusters with eps={eps}')
    plt.show()
```

```python
from sklearn.cluster import DBSCAN 

# Your code here

# Create the DBSCAN model with eps=0.5
dbscan = DBSCAN(eps=0.5)

# Fit the model to the scaled data (excluding the KMeans label column)
dbscan.fit(customers_scale.drop(columns='Label'))

# Assign the DBSCAN labels back to the DataFrame
customers_scale['labels_DBSCAN'] = dbscan.labels_

# Display the first rows with the new DBSCAN cluster labels
print(customers_scale.head())
```
